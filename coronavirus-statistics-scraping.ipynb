```python
url = 'https://www.worldometers.info/coronavirus/'
source  = requests.get(url).text
soup = bs4.BeautifulSoup(source,'html.parser')
```

```python
headings = [head.text.strip().replace('\n','') for head in soup.find_all('th')[1:15]]
headings
```

```python
pd.DataFrame(columns=headings)
```

```python
datas = soup.find_all('td')
datas[176:]
```

looking at the td , we find that , row has data for 22 columns , so to process this and turn this rowdata of 22columns into data for 14columns 
first we need to make a nested list and group 22row data into each list ,
after that grouping the first 14 row data and appending it to our row_data , (also checked for an empty element and replaced them with "N/A" for better understanding)

```python
def scrape(soup,headings):
    datas = soup.find_all('td');
    td_elements = [element.text for element in datas[176:]]


    row_data = []
    for i in range(0, len(td_elements), 22):  
      group = td_elements[i:i+22]  
      processed_group = []
      for element in group[:14]:  
        processed_group.append("N/A" if element == '' else element)  
      row_data.append(processed_group)  


    df = pd.DataFrame(columns=headings, data=row_data)
    return df
    
```
